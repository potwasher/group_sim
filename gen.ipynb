```python
from typing import List
import numpy as np
import matplotlib.pyplot as plt
```

```python
PLAYER_CNT = 2000
internal_ratings: List[int] = []
rng = np.random.default_rng()
for _ in range(PLAYER_CNT):
    internal_ratings.append(round(rng.normal(1500, 150)))
internal_ratings.sort()
plt.hist(internal_ratings)
```

```python
f = open('ratings.txt', "w")
print(PLAYER_CNT, file=f)
print(*internal_ratings, file=f)
```
